```python
import re
import pandas as pd
import string
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
## Importing data csv file

df = pd.read_csv('Musical_instruments_reviews.csv')

df.reviewText
```

```python
## Cleaning review section

## Removing capital letters

df.reviewText = df.reviewText.str.lower()

## Removing punctuation
df.reviewText = df.reviewText.map(lambda x: re.sub('[%s]' % re.escape(string.punctuation), ' ', str(x)))
df
```

```python
review = list(df.reviewText)

cv = CountVectorizer(stop_words = 'english', min_df = 0.0005)
x = cv.fit_transform(review)
review_matrix = pd.DataFrame(x.toarray(), columns=cv.get_feature_names())

review_matrix
```

```python
## data acqusition turning in part

words = list(review_matrix.columns)
len(words)
```

```python
cosine_matrix = cosine_similarity(review_matrix)
cosine_matrix
```

```python
def get_largest_cosine(cosine_array):
    largest_cosine = 0
    for lst in cosine_array:
        for num in lst:
            if num <= .999 and num >= largest_cosine:
                largest_cosine = num
    return largest_cosine

get_largest_cosine(cosine_matrix)
```

```python
largest_cosine = get_largest_cosine(cosine_matrix)
largest_cosine
```

```python
## sentiment test of summary

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

df['score'] = df.summary.map(analyzer.polarity_scores).map(lambda x: x.get('compound'))

df['score'].head(3)
```
